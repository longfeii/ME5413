```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN,OPTICS
from sklearn import metrics
import seaborn as sns
import pandas as pd
file_data = np.fromfile('me5413_homework1/me5413/1_lidar/lidar_data/frame7.pcd.bin', dtype=np.float32)
#reshape the data
points = file_data.reshape((-1, 5))[:, :4]
#adjust parameter
result= []#deposit noise ratio and score of different parameters
eps = np.arange(3,8,0.5) #eps from 0.1 to 8 with interval 0.4
min_samples=np.arange(10,50,5)#min_samples from 2 to 50

best_score=0
best_score_eps=0
best_score_min_samples=0

for i in eps:
 for j in min_samples:
     try:
                cluster = OPTICS(max_eps=np.inf,min_samples=j, cluster_method='dbscan',eps=i).fit(points)
                labels= cluster.labels_
                k=metrics.silhouette_score(points,labels) #Silhouette coefficient
                raito = len(labels[labels[:] == -1]) / len(labels) #noise ratio
                n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
                result.append([i,j,k,raito,n_clusters_])
                if k > best_score:
                    best_score = k
                    best_score_eps = i
                    best_score_min_samples = j

                else:
                    cluster = ''

     except:
                cluster='' #if report an error

result= pd.DataFrame(result)
result.columns=['eps','min_samples','score','raito','n_clusters']
sns.relplot(x="eps",y="min_samples", size='score',data=result)
sns.relplot(x="eps",y="min_samples", size='raito',data=result)
print(best_score_eps)
print(best_score_min_samples)
plt.show()
# import numpy as np
# import matplotlib.pyplot as plt
# from sklearn.cluster import DBSCAN,OPTICS,AgglomerativeClustering
# from sklearn import metrics
# import seaborn as sns
# import pandas as pd
# file_data = np.fromfile('me5413_homework1/me5413/1_lidar/lidar_data/frame1.pcd.bin', dtype=np.float32)
# #reshape the data
# points = file_data.reshape((-1, 5))[:, :4]
# #adjust parameter
# result= []#deposit noise ratio and score of different parameters
# cluster_n = np.arange(3,30,3) #eps from 0.1 to 8 with interval 0.4
# sample_way=['ward', 'complete', 'average', 'single']#min_samples from 2 to 50
#
# best_score=0
# best_score_cluster_n=0
# best_score_sample_way=0
#
# for i in cluster_n:
#  for j in sample_way:
#      try:
#                 cluster = AgglomerativeClustering(linkage=j, n_clusters=i).fit(points)
#                 labels= cluster.labels_
#                 k=metrics.silhouette_score(points,labels) #Silhouette coefficient
#                 raito = len(labels[labels[:] == -1]) / len(labels) #noise ratio
#                 n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
#                 result.append([i,j,k,raito,n_clusters_])
#                 if k > best_score:
#                     best_score = k
#                     best_score_cluster_n = i
#                     best_score_sample_way = j
#
#                 else:
#                     cluster = ''
#
#      except:
#                 cluster='' #if report an error
#
# result= pd.DataFrame(result)
# result.columns=['cluster_n','sample_way','score','raito','n_clusters']
# sns.relplot(x="cluster_n",y="sample_way", size='score',data=result)
# sns.relplot(x="cluster_n",y="sample_way", size='raito',data=result)
# print(best_score_cluster_n)
# print(best_score_sample_way)
# plt.show()
```
